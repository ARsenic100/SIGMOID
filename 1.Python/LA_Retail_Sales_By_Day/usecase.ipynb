```python
import pandas as pd
import os
```

Top 3 stores having highest revenue in the month

```python
files = []
for day in range(1, 31):
    file_path = f"sales_2024-09-{str(day).zfill(2)}.csv"
    if os.path.exists(file_path):
        files.append(pd.read_csv(file_path))
    else:
        print(f"Missing: {file_path}")
```

```python
data = pd.concat(files, ignore_index=True)
print(data.head())
```

```python
data["month"] = data["date"].astype(str).str[:7]
data[["date", "month"]].head()
```

```python
top_stores = data.groupby(['month','store_id', 'store_name'])['dollar_sales'].sum().nlargest(3)
print(top_stores)
```
