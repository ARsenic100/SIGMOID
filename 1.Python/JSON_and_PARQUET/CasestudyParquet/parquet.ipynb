Retrieve top 10 rows by starr rating from Parquet file: customer_reviews_with_sentiment.parquet

```python
import pandas as pd
```

```python
pip install fastparquet
```

```python
reviews_df = pd.read_parquet("customer_reviews_with_sentiment.parquet", engine="fastparquet")
reviews_df.head()
```

```python

reviews_df.head(10)
```

```python
top10_by_star = (
    reviews_df[["product_title", "star_rating", "sentiment", "review_headline", "review_body"]]
      .sort_values("star_rating", ascending=False)
      .head(10)
      .reset_index(drop=True)
)

top10_by_star
```
