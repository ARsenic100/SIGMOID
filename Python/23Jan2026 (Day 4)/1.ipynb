SINGLE INHERITANCE

```python
class Apple:
    #Class attributes
    manufacturer= 'Apple Inc.'
    contactWebsite= 'https://www.apple.com/in/apple-watch-series-11/'

    #Method
    def contactDetails(self):
        print('To contact us, visit:' , self.contactWebsite)
```

```python
class MacBook(Apple):
    def __init__(self):
        self.yearOfManufacture = 2020
    def manufacturerDetails(self):
        print(f"This macbook was manufactured in the year {self.yearOfManufacture}")
        print(f"This macbook was manufactured by {self.manufacturer}")
```

```python
macBook = MacBook()
macBook.contactDetails()
macBook.manufacturerDetails()
```
