```python
#Creating a JSON string with escaped double quotes

order_jd = "{\"order_id\": 1, \"order_date\": \"2013-07-25\"}"

order_jd
     
#Importing the json module

import json
     
#Converting JSON string into a Python dictionary

order = json.loads(order_jd)

order
     
#Checking the type of the converted object

type(order)
     
#Accessing a value from the dictionary

order['order_date']
     
#Creating a JSON string using single quotes

order_jd = '{"order_id": 1, "order_date": "2013-07-25"}'
     
#Parsing JSON and accessing a field directly

json.loads(order_jd)['order_date']
```
