```python
filePath = 'samplefile.txt'
```

```python
print(filePath)
```

```python
linesRdd = spark.sparkContext.textFile(filePath)
```

```python
linesRdd.collect()
```

```python
linesRdd.map(lambda x: x.split()).collect()
```

```python
wordsRdd= linesRdd.flatMap(lambda x: x.split())
```

```python
wordsMapRdd= wordsRdd.map(lambda x: (x, 1))
```

```python
wordCountRdd = wordsMapRdd.reduceByKey(lambda a, b: a + b)
```

```python
wordCountRdd.collect()
```
