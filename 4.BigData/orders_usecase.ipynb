```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()


ordersDataPath = 'orders.txt'

ordersRdd = spark.sparkContext.textFile(ordersDataPath)
```

```python
ordersMapRdd = ordersRdd.map(lambda x: (x.split(',')[3], 1))
```

```python
ordersStatusRdd = ordersMapRdd.reduceByKey(lambda a, b: a + b)
```
