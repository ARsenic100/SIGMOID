```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Colab-PySpark").getOrCreate()
```

```python
filePath= 'access.log.1'
logRdd = spark.sparkContext.textFile(filePath)
```

```python
for i in logRdd.take(10):
  print(i)
```

```python
x= logRdd.first()
x
```

```python
x.split(' ')
```

```python
x.split(' ')[8]
```

```python
logRdd.map(lambda x: x.split(' ')[8]).collect()
```

```python
logRdd.filter(lambda x: x.split()[8] in ['400', '404','500']).count()
```

```python
logRdd \
.filter(lambda x: x.split()[8] in ['400', '404','500']) \
.map(lambda x: (x.split()[8], 1)) \
.reduceByKey(lambda x, y: x + y) \
.collect()
```
